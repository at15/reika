```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt
```

Hours spend on course project per day

```python
weeks = [4, 5, 6, 7, 8, 9, 10]
hours = [0.5, 0.3, 0.2, 0.1, 1.0, 6.0, 0.3]
plt.plot(weeks, hours)
plt.ylabel('hours spend on course project per day')
plt.xlabel('week')
plt.show()
```

```python
locs = [100, 110, 110, 120, 500, 2000, 2000]
plt.plot(weeks, locs)
plt.ylabel('lines of code in course project')
plt.xlabel('week')
plt.show()
```
